# $\bf{1D\ diffusion:\ }$ $\frac{\partial u}{\partial t} = \nu \frac{\partial^2 u}{\partial x^2}$

```python
# LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# SPACE GRID
NX = 101
XINI = 0.0
XEND = 2.0
DX = (XEND - XINI)/(NX - 1)
X = np.linspace(XINI, XEND, NX)

# TIME PARAMS
NT = 4000
DT = 0.00001
TINI = 0.0
TEND = TINI + NT * DT

# PROBLEM PARAMS
VIS = 4.0    # VISCOSITY = -0.1, FOR EXPLOSION

# BOUNDARY CONDITIONS
ULEFT = 1.0
URIGHT = 1.0
```

```python
# INITIAL CONDITION
U = np.zeros(X.shape)
UN = np.zeros(U.shape)
UINI = np.zeros(U.shape)

for i in np.arange(0, len(X)):
    if X[i] >= 0.5 and X[i] <= 1.0:
        U[i] = 2.0
        UN[i] = 2.0
        UINI[i] = 2.0
    else:
        U[i] = 1.0
        UN[i] = 1.0
        UINI[i] = 1.0
```

```python
# FORWARD DIFF. IN TIME, CENTRAL DIFF. IN SPACE
for j in range(NT):
    U[0] = ULEFT
    U[-1] = URIGHT
    UN = U.copy()

    for i in np.arange(1, NX-1):
        U[i] = UN[i] + VIS * (DT/DX**2) * (UN[i+1] - 2*UN[i] + UN[i-1])

print(U)
```

```python
# ANALYTICAL SOLUTION ~ FOURIER

# SPACE GRID
NX = 101
XINI = 0.0
XEND = 2.0
DX = (XEND - XINI)/(NX - 1)
X = np.linspace(XINI, XEND, NX)

# TIME PARAMS
NT = 4000
DT = 0.00001
TINI = 0.0
TEND = TINI + NT * DT

# PROBLEM PARAMS
VIS = 4.0    # VISCOSITY = -0.1, FOR EXPLOSION
```

```python
# COMPUTE FOURIER COEFFICIENTS
NTERMS = 200
A = np.zeros(NTERMS)

for n in range(NTERMS):
    A[n] = 2.0/((n + 1) * np.pi) * (np.cos(np.pi/4.0 * (n + 1)) - np.cos(np.pi/2.0 * (n + 1)))
```

```python
# COMPUTE THE ANALYTICAL SOLUTION
UA = np.zeros(X.shape)

T = TEND

for i in range(NX):
    UA[i] = 1.0
    
    for n in range(NTERMS):
        UA[i] = UA[i] + A[n] * np.sin((n + 1)/2.0 * np.pi * X[i]) * np.exp(-(n + 1)**2/4.0 * np.pi**2 * VIS * T)
        
print(UA)    
```
