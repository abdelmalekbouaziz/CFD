# $\textbf{1D\ linear\ convection:\ }$ $\frac{\partial u}{\partial t} + c \frac{\partial u}{\partial x} = 0$

```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Space Grid
nx = 201
x_b = 0.0
x_e = 2.0
u_b = 1.0
u_e = 1.0
dx = (x_e - x_b)/(nx - 1)
x = np.linspace(x_b, x_e, nx)

# Time Grid
nt = 51 * 2
dt = 0.01
t_b = 0.0
t_e = nt * dt
t = np.linspace(t_b, t_e, nt)

# Equation Constants
c = 1.0     # c < 1.0 for artificial viscosity effects ~ 0.5
```

```python
# Initial Solution
u = np.zeros(x.shape)
un = np.zeros(x.shape)

for i in np.arange(0, len(x)):

    if x[i] >= 0.5 and x[i] <= 1.0:
        u[i] = 2.0

    else:
        u[i] = 1.0
```

```python
# Forward Difference in Time | Backward Difference in Space Scheme
for i in range(len(t)):

    # Dirichlet boundary conditions
    u[0] = u_b      # Left boundary
    u[-1] = u_e     # Right boundary

    un = u.copy()
    
    for j in range(1, len(x)-1):
        u[j] = un[j] - c * (dt / dx) * (un[j] - un[j-1])

print(u)
```
